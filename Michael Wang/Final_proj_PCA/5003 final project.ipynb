```python
plt.figure(figsize=(12,10))
k=1
# for i in range1:
#     plt.scatter(df[1].loc[i],df[2].loc[i],label=df['Unnamed: 0'].loc[i],color='red')
plt.scatter(df[k].loc[range1],df[k+1].loc[range1],color='red',label='Basic materials')
# for i in range2:
plt.scatter(df[k].loc[range2],df[k+1].loc[range2],color='black',label='Consumer goods')
# for i in range3:
plt.scatter(df[k].loc[range3],df[k+1].loc[range3],color='green',label='Financials')
# for i in range4:
plt.scatter(df[k].loc[range4],df[k+1].loc[range4],color='blue',label='Health care')
# for i in range5:
plt.scatter(df[k].loc[range5],df[k+1].loc[range5],color='orange',label='Industrials')
# for i in range6:
plt.scatter(df[k].loc[range6],df[k+1].loc[range6],color='yellow',label='Oil and gas')
# for i in range7:
plt.scatter(df[k].loc[range7],df[k+1].loc[range7],color='purple',label='Technology')
# for i in range8:
plt.scatter(df[k].loc[range8],df[k+1].loc[range8],color='darkseagreen',label='Telecommunications')
# for i in range9:
plt.scatter(df[k].loc[range9],df[k+1].loc[range9],color='skyblue',label='Utilities')


plt.axhline(0,linestyle='--')
plt.axvline(0,linestyle='--')

# plt.text(-0.1,0,'$\\star$',fontsize=50,color='red')
plt.legend()
plt.title('Component %d verus component %d'%(k+1,k))
plt.xlabel('component %d'%k)
plt.ylabel('component %d'%(k+1))
plt.savefig('Component %d verus component %d.png'%(k+1,k))
```
