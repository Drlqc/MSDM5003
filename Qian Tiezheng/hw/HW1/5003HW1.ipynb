```python
import numpy as np
import math

#import matplotlib.mlab as mlab

from scipy.stats import norm
from numpy.linalg import cholesky
import matplotlib.pyplot as plt
sampleNo=10000
mu1=0
mu2=1
dev1=1
dev2=2
#x=np.linspace
np.random.seed(50)
s1=np.random.normal(mu1,dev1,sampleNo)
s2=np.random.normal(mu2,math.sqrt(dev2),sampleNo)
z=s1+s2
(mu,sigma)=norm.fit(z)
print('fitting result','\nmu=',mu,'\ndeviation=',sigma*sigma)
plt.hist(s1,50,density=True,color='y')
plt.hist(s2,50,density=True,color='k')
n,bins,patches = plt.hist(z,50,density=True,color='g')
z_simu=norm.pdf(bins,mu,sigma)
z_real=norm.pdf(bins,1,math.sqrt(3))
x=norm.pdf(bins,0,1)
y=norm.pdf(bins,1,math.sqrt(2))
plt.plot(bins,x,linewidth=2.0,color='m',label='X')
plt.plot(bins,y,linewidth=2.0,color='c',label='Y')
plt.plot(bins,z_simu,linewidth=2.0,color='r',linestyle='--',label='Z-fit')
plt.plot(bins,z_real,linewidth=2.0,color='b',label='Z-theory')
plt.title('Li,Qichao 5003HW1\nThe pdf of the sum of 2 normal random variables')
plt.legend(loc='best')
plt.ylabel('PDF(x)',fontsize=15)

plt.xlabel('x',fontsize=15)
plt.show()
```
